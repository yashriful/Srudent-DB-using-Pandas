## loc and iloc operators

```python
df.head()
```

```python
df.loc[0, "name"] # first row and name column
# loc[rows, columns]
```

```python
# return mark for Krish
df.loc[3, "mark"]
```

```python
df.loc[3, ["name", "mark"]]
```

```python
df.loc[0:3, ["name", "mark"]] #loc operator return stopping points included
```

```python
df.head()
```

```python
# return mark for Krish
df.iloc[3, 3] # iloc[rows, columns]
```

```python
df.iloc[0:3, [1,2]] #iloc doesnt include stopping point on range
```

```python
df.loc[0:2, ["name", "class"]]
```

```python
df.iloc[1:3 , 1] # for row range stropping point (index 3) will not be in the output
```

```python
df.iloc[1:3 , [1,3]] # for row range stropping point (index 3) will not be in the output
```

```python
# What is the index for gender column ?
df.columns.get_loc("gender")
```

```python
df.iloc[1:3 , [1, df.columns.get_loc("gender")]] # for row range stopping point (index 3) will not be in the output
```
